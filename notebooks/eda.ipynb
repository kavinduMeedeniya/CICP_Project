```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Load data
df = pd.read_csv('../data/raw/Car_Insurance_Claim.csv')

# Create output directory
os.makedirs('../outputs/figures', exist_ok=True)

# Print dataset info
print("Dataset Info:")
df.info()
print("\nMissing Values:")
print(df.isnull().sum())

# Graphs
plt.figure(figsize=(6, 4))
sns.countplot(x='OUTCOME', data=df)
plt.title('Claim Distribution')
plt.savefig('../outputs/figures/claim_distribution.png')
plt.show()
plt.close()

numerical_cols = ['CREDIT_SCORE', 'ANNUAL_MILEAGE', 'SPEEDING_VIOLATIONS', 'DUIS', 'PAST_ACCIDENTS', 'OUTCOME']
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('../outputs/figures/correlation_heatmap.png')
plt.show()
plt.close()

plt.figure(figsize=(8, 6))
sns.countplot(x='DRIVING_EXPERIENCE', hue='OUTCOME', data=df)
plt.title('Claims by Driving Experience')
plt.savefig('../outputs/figures/claims_by_experience.png')
plt.show()
plt.close()
```
